```python
# Assume openai>=1.0.0
def get_llm_prediction(api_key,base_url,model,prompt,text):
    from openai import OpenAI

    # Create an OpenAI client with your deepinfra token and endpoint
    openai = OpenAI(
        api_key=api_key,
        base_url=base_url,
    )

    chat_completion = openai.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt+text}],
    )

    return chat_completion.choices[0].message.content


```

```python
import pandas as pd
import numpy as np
def read_csv_column(file_path, column_name):
    df = pd.read_csv(file_path)
    
    column_data = df[column_name].replace('', np.nan).fillna('null')
    
    return column_data.to_list()
```

```python
api_key=""
base_url=""
model=""

file_path="../data/Traffic/US_VMT_Month.csv"
column_name="Final_Search_4"
text_list=read_csv_column(file_path, column_name)
prompt="As an expert in the field of traffic with a focus on Travel Volulmn forecasting, you will be provided with textual information about Travel Volulmn in the United States from a past period. Based on this text and your expertise, please make a concise prediction about the Travel Volulmn situation in the United States for the upcoming period. Your description of prediction should be brief and expressed in one sentence. The input text information is as follows:"
```

```python
get_llm_prediction(api_key,base_url,model,prompt,text_list[3])
```

```python
predtext_list=[]
for num,text in enumerate(text_list):
    
    print(num/len(text_list))
    predtext_list.append(get_llm_prediction(api_key,base_url,model,prompt,text))
```

```python
import csv
import os
def update_csv_with_list(csv_path, attribute_name, string_list):
    rows = []
    headers = []
    try:
        with open(csv_path, 'r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            headers = reader.fieldnames
            rows = list(reader)
    except FileNotFoundError:
        print(f"CSV file not found: {csv_path}")
        return
    
    if attribute_name not in headers:
        headers.append(attribute_name)
    
    if len(string_list) > len(rows):
        print(f"Warning: List length ({len(string_list)}) exceeds CSV row count ({len(rows)}). Extra items will be ignored.")
    
    for i, row in enumerate(rows):
        if i < len(string_list):
            row[attribute_name] = string_list[i]
        else:
            row[attribute_name] = ''
    
    with open(csv_path, 'w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=headers)
        writer.writeheader()
        writer.writerows(rows)
    
    print(f"CSV file updated successfully: {csv_path}")
```

```python
update_csv_with_list(csv_path=file_path, attribute_name="Final_Output", string_list=predtext_list)
```
